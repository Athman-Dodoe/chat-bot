```python
from transformers import pipeline, AutoTokenizer
import gradio as gr
```

```python
activity_dataset = {
    "POSITIVE": {
        "joy":[
            "Go for a walk in nature",
            "Try a new hobby",
            "Call a friend to share your joy"
        ],

        "love": [
            "Write a gratitude letter",
            "Spend time with loved ones",
            "Do something kind for someone"
        ],
        "surprise": [
            "Try something spontaneous",
            "Visit a new place",
            "Do something out of your routine"
        ],
        "happy": [
            "Listen to uplifting music",
            "Dance to your favorite songs",
            "Treat yourself to something nice"
        ]
    },
    # Negative sentiment activities
    "NEGATIVE": {
        "sadness": [
            "Write in a journal",
            "Listen to calming music",
            "Practice deep breathing exercises"
            ],
        "anger": [
            "Do intense physical exercise",
            "Try punching a pillow",
            "Write down your feelings then tear it up"
            ],
        "fear": ["Practice mindfulness meditation",
                 "Make a plan to address your fears",
                 "Talk to a trusted friend"
                 ],
        "anxiety": [
            "Do progressive muscle relaxation",
            "Try a grounding exercise",
            "Make a comforting routine"
            ]
    }
}
```

```python

sentiment_model_name = "distilbert-base-uncased-finetuned-sst-2-english"
sentiment_tokenizer = AutoTokenizer.from_pretrained(sentiment_model_name)
sentiment_model = pipeline("sentiment-analysis", model=sentiment_model_name, tokenizer=sentiment_tokenizer)

emotion_model_name = "bhadresh-savani/distilbert-base-uncased-emotion"
emotion_tokenizer = AutoTokenizer.from_pretrained(emotion_model_name)
emotion_model = pipeline("text-classification", model=emotion_model_name, tokenizer=emotion_tokenizer)

def analyze_mood_and_suggest_activities(message, history):
  """Process user message and return activity suggestions"""
  sentiment_result = sentiment_model(message)[0]
  emotion_result = emotion_model(message)[0]

  sentiment = sentiment_result['label'].upper()
  emotion = emotion_result['label'].lower()

  try:
    recommended_activities = activity_dataset[sentiment][emotion]
  except KeyError:
    try:
      recommended_activities = activity_dataset[sentiment]['general']
    except KeyError:
      recommended_activities = ["Take a deep breath", "Go for a short walk", "Drink some water"]

  response = (
    f"Sentiment: {sentiment_result['label']}\n"
    f"Emotion: {emotion_result['label']}\n\n"
    "Here are some activities that might help:\n"
  )
  for i, activity in enumerate(recommended_activities, 1):
      response += f"{i}. {activity}\n"

  return response


```

```python
chatbot = gr.ChatInterface(
    fn = analyze_mood_and_suggest_activities,
    title = "Mood-Based Activity Recommender",
    description = '<p style="text-align: center;">Share how you\'re feeling, and I\'ll suggest helpful activities for you!</p>',
    examples = ["I'm feeling great today!", "I'm so angry right now", "I feel anxious about tomorrow"],
    theme = "soft",
)
chatbot.launch(pwa=True)
```
